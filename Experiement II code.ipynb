## Create New timestamp features

```python
# Add year-month occurrence count
month_year = (df_all.timestamp.dt.month + df_all.timestamp.dt.year * 100)
month_year_cnt_map = month_year.value_counts().to_dict()
df_all['month_year_cnt'] = month_year.map(month_year_cnt_map)
```

```python
# Add year-week occurrence count
week_year = (df_all.timestamp.dt.weekofyear + df_all.timestamp.dt.year * 100)
week_year_cnt_map = week_year.value_counts().to_dict()
df_all['week_year_cnt'] = week_year.map(week_year_cnt_map)
```

```python
# Add month and day-of-week
df_all['month'] = df_all.timestamp.dt.month
df_all['dow'] = df_all.timestamp.dt.dayofweek
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
numerical_transformer = SimpleImputer(strategy='median')
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
sub_area_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('label',OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])
```

```python
from sklearn.tree import DecisionTreeRegressor
model1 = DecisionTreeRegressor(random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor(random_state=1)
```

```python
from xgboost import XGBRegressor
model3 = XGBRegressor(random_state=1,early_stopping_rounds=5)
```

```python
top20_features = ['full_sq',
 'sub_area',
 '0_13_female',
 '7_14_female',
 'work_female',
 '0_13_all',
 '0_17_male',
 'young_female',
 '0_17_female',
 'raion_popul',
 '7_14_male',
 '0_13_male',
 'work_male',
 'work_all',
 'young_all',
 '0_17_all',
 'children_school',
 '0_6_female',
 'ekder_female',
 '0_6_male']
```

```python
times_new_features = ['month_year_cnt', 'week_year_cnt', 'month', 'dow']
```

```python
macro_cols = ["balance_trade", "balance_trade_growth", "eurrub", "average_provision_of_build_contract",
"micex_rgbi_tr", "micex_cbi_tr", "deposits_rate", "mortgage_value", "mortgage_rate",
"income_per_cap", "rent_price_4+room_bus", "museum_visitis_per_100_cap", "apartment_build"]
```

```python
combined_features = top20_features + times_new_features + macro_cols
```

```python
X_combined = X_uesful_train[combined_features]
```

```python
numerical_cols_combined = ['full_sq', '0_13_female', '7_14_female', 'work_female',
       '0_13_all', '0_17_male', 'young_female', '0_17_female', 'raion_popul',
       '7_14_male', '0_13_male', 'work_male', 'work_all', 'young_all',
       '0_17_all', 'children_school', '0_6_female', 'ekder_female', '0_6_male',
       'month_year_cnt', 'week_year_cnt', 'month', 'dow', 'balance_trade',
       'balance_trade_growth', 'eurrub', 'average_provision_of_build_contract',
       'micex_rgbi_tr', 'micex_cbi_tr', 'deposits_rate', 'mortgage_value',
       'mortgage_rate', 'income_per_cap', 'rent_price_4+room_bus',
       'museum_visitis_per_100_cap', 'apartment_build']
```

```python
categorical_cols_combined = ['sub_area']
```

```python
preprocessor_combined = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols_combined),
        ('sub_area', sub_area_transformer, categorical_cols_combined)
    ])
```

```python
my_pipeline1_combined = Pipeline(steps=[
    ('preprocessor', preprocessor_combined),
    ('model', model1),
])
```

```python
my_pipeline2_combined = Pipeline(steps=[
    ('preprocessor', preprocessor_combined),
    ('model', model2),
])
```

```python
my_pipeline3_combined = Pipeline(steps=[
    ('preprocessor', preprocessor_combined),
    ('model', model3),
])
```

```python
from sklearn.model_selection import GridSearchCV
```
